```python
import pandas as pd
import numpy as np
```

```python
filepath="covid-chestxray-dataset-master/metadata.csv"
imagespath="covid-chestxray-dataset-master/images"
```

```python
df=pd.read_csv(filepath)
print(df.shape)
```

```python
df.head()
```

```python
import os
import shutil
```

```python
targetdir="NewDataset/Covid"
if not os.path.exists(targetdir):
    os.mkdir(targetdir)
    print("covid folder created")
```

```python
for(i,row) in df.iterrows():
    print(i,row)
```

```python
count=0
for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19":
        count+=1
print("no of covid images ",count)
```

```python
count=0
for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        count+=1
print("no of covid images that are PA ",count)
```

```python
count=0
for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        curr_imagepath=os.path.join(imagespath,filename)
        image_copy_path=os.path.join(targetdir,filename)
        shutil.copy2(curr_imagepath,image_copy_path)
        count+=1
print("no of covid images ",count)
```

```python
import random
kaggle_file_path="chest_xray_kaggle/train/NORMAL"
target_normal_dir="NewDataset/Normal"
```

```python
image_names=os.listdir(kaggle_file_path)
```

```python
random.shuffle(image_names)
```

```python
for i in range(196):
    image_name=image_names[i]
    image_path=os.path.join(kaggle_file_path,image_name)
    target_path=os.path.join(target_normal_dir,image_name)
    
    shutil.copy2(image_path,target_path)
    print("copying image ",i)
```
